```python
#install required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotnine import *
```

```python
#load the data
df = pd.read_csv("/content/drive/My Drive/WORK/SI_Budget_Report/fy24-capital-budget-plan-recommended.csv")

df.head()
```

```python
df.shape
```

```python
# Clean column Total_Project_Budget
df["Total_Project_Budget"] = pd.to_numeric(df["Total_Project_Budget"].astype(str).str.replace(",", ""), errors='coerce').astype('Int64')
```

```python
# Strip whitespaces from column names
df.columns = df.columns.str.strip()
```

```python
# Drop column Scope_Of_Work	and Project_Name
df.drop(["Scope_Of_Work", "Project_Name"], axis=1, inplace=True)
```

```python
# Replace all '0' with NaN (null values)
df.replace(0, np.nan, inplace=True)
```

```python
df.info()
```

```python
columns_to_drop = ['Authorization_Future', 'Grant_Existing', 'Grant_FY', 'Grant_Future', 'Grant_Expended', 'Grant_Year_0', 'Grant_Year_1', 'GrantYear_25', 'External_Funds']
df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
df.head()
```

```python
## Look over these values to determine which values to group together
department_counts = df["Department"].value_counts()
pm_department_counts = df["PM_Department"].value_counts()

print("Department counts:")
print(department_counts)

print("\nPM_Department counts:")
print(pm_department_counts)
```

```python
## Export ungrouped departments
## Export to csv file
df.to_csv('/content/drive/My Drive/WORK/SI_Budget_Report/clean_CapBudget.csv', index=False)
```

```python
# Group together similar departments

## Group departments Fire Department, Police Department
## Group departments Environment Department, Parks and Recreation Department

## Office of New Urban Mechanics into Department of Innovation and Technology
## Repairs and Rennovations (Property Management Department, Public Health Commission)
df['Department'] = df['Department'].replace({
    'Fire Department': 'Police/Fire Departments',
    'Police Department': 'Police/Fire Departments',
    'Environment Department': 'Environmental Departments',
    'Parks and Recreation Department': 'Environmental Departments',
    'Property Management Department': 'Repairs and Renovation Departments',
    'Public Health Commission': 'Repairs and Renovation Departments',
    'Office of New Urban Mechanics': 'Department of Innovation and Technology'
})
```

```python
# Group together similar PM_departments

## Group departments Boston Fire Department, Boston Police Department, Police Department
## Group departments Environment Department, Parks and Recreation Department
## Group departments Boston Centers for Youth and Families and Youth Engagement and Employment

## Office of New Urban Mechanics into Department of Innovation and Technology
## Property Management Department and Boston Public Health Commission
## and Boston Public Library into Public Facilities Department
## Mayor's Office of Housing into Boston Housing Authority
df['PM_Department'] = df['PM_Department'].replace({
    'Boston Fire Department': 'Police/Fire Departments',
    'Boston Police Department': 'Police/Fire Departments',
    'Police Department': 'Police/Fire Departments',
    'Environment Department': 'Environmental Departments',
    'Parks and Recreation Department': 'Environmental Departments',
    'Boston Centers for Youth and Families': 'Youth Departments',
    'Youth Engagement and Employment': 'Youth Departments',
    'Office of New Urban Mechanics': 'Department of Innovation and Technology',
    'Property Management Department': 'Public Facilities Department',
    'Boston Public Health Commission': 'Public Facilities Department',
    'Boston Public Library': 'Public Facilities Department',
    "Mayor's Office of Housing": 'Boston Housing Authority'
})
```

```python
## Look over these values to determine which values to group together
department_counts = df["Department"].value_counts()
pm_department_counts = df["PM_Department"].value_counts()

print("Department counts:")
print(department_counts)

print("\nPM_Department counts:")
print(pm_department_counts)
```
